```python
from tkinter import ttk,messagebox
from PIL import Image,ImageTk
import pymysql
```

```python
class register:
    def __init__(self,appu):
        self.appu = appu
        self.appu.title("REGISTRATION FORM")
        self.appu.geometry("1350x700+0+0")
        self.appu.configure(bg="white")
        
        self.b2=ImageTk.PhotoImage(file="bg.png")
        b2= Label(self.appu,image=self.b2).place(x=0,y=0,relwidth=1,relheight=1)
        
        
        self.lef=ImageTk.PhotoImage(file="event.jpg")
        lef= Label(self.appu,image=self.lef).place(x=80,y=100,width=400,height=500)
        
        frame = Frame(self.appu,bg="#E5E4E2")
        frame.place(x=480,y=100,width=700,height=500)
        
        title = Label(frame,text="Registration here",font=("time new roman",20,"bold"),bg="white",fg="green").place(x=50,y=30)
        
        
        #rr
        
        fname = Label(frame,text="First Name",font=("time new roman",15,"bold"),bg="white",fg="gray").place(x=50,y=100)
        self.txt_frame1 = Entry(frame,font=("time new roman",15),bg="white")
        self.txt_frame1.place(x=50,y=130,width=250)
        
        
        lname = Label(frame,text="Last Name",font=("time new roman",15,"bold"),bg="white",fg="gray").place(x=370,y=100)
        self.txt_frame2 = Entry(frame,font=("time new roman",15),bg="white")
        self.txt_frame2.place(x=370,y=130,width=250)
       
        
        #rrr
        contactname = Label(frame,text="Contacts No",font=("time new roman",15,"bold"),bg="white",fg="gray").place(x=50,y=170)
        self.txt_frame3 = Entry(frame,font=("time new roman",15),bg="white")
        self.txt_frame3.place(x=50,y=200,width=250)
       
    
        emailname = Label(frame,text="Email Address",font=("time new roman",15,"bold"),bg="white",fg="gray").place(x=370,y=170)
        self.txt_frame4 = Entry(frame,font=("time new roman",15),bg="white")
        self.txt_frame4.place(x=370,y=200,width=250)
        
        #rrrr
        questionname = Label(frame,text="Security Question",font=("time new roman",15,"bold"),bg="white",fg="gray").place(x=50,y=240)
        self.txt_frame5 = ttk.Combobox(frame,font=("time new roman",13),state="readonly",justify = CENTER)
        self.txt_frame5["value"]=("--Select--","Your first pet name","Your birth place","Your best friend name")
        self.txt_frame5.place(x=50,y=270,width=250)
        self.txt_frame5.current(0)
        

        
        answername = Label(frame,text="Answer",font=("time new roman",15,"bold"),bg="white",fg="gray").place(x=370,y=240)
        self.txt_frame6 = Entry(frame,font=("time new roman",15),bg="white")
        self.txt_frame6.place(x=370,y=270,width=250)
        
        
        #rr
        passwordname = Label(frame,text="Password",font=("time new roman",15,"bold"),bg="white",fg="gray").place(x=50,y=310)
        self.txt_frame7 = Entry(frame,font=("time new roman",15),bg="white")
        self.txt_frame7.place(x=50,y=340,width=250)
        
        
        emailname = Label(frame,text="Confirm Password",font=("time new roman",15,"bold"),bg="white",fg="gray").place(x=370,y=310)
        self.txt_frame8 = Entry(frame,font=("time new roman",15),bg="white")
        self.txt_frame8.place(x=370,y=340,width=250)
        
        #tt
        self.chk_check = IntVar()
        chk =Checkbutton(frame,text="I agree the term and conditions",variable=self.chk_check ,onvalue=1,offvalue=0,bg="white",font=("time new roman",12)).place(x=50,y=380)
        
        but1 = Button(frame,text="Register",bg="green",font=("time new roman",12),command=self.register_data,width = 65).place(x=50,y=420)
        but2 = Button(frame,text="Sign in",bg="green",font=("time new roman",12),width = 65).place(x=50,y=470)
        
    def clear(self):
        self.txt_frame1.delete(0,END)
        self.txt_frame2.delete(0,END)
        self.txt_frame3.delete(0,END)
        self.txt_frame4.delete(0,END)
        self.txt_frame5.delete(0,END)
        self.txt_frame6.delete(0,END)
        self.txt_frame7.delete(0,END)
        self.txt_frame8.delete(0,END)


    def register_data(self):
        if self.txt_frame1.get()=="" or self.txt_frame2.get()=="" or self.txt_frame3.get()=="" or self.txt_frame4.get()=="" or self.txt_frame5.get()=="" or self.txt_frame6.get()=="" or self.txt_frame7.get()=="" or self.txt_frame8.get()=="":
            messagebox.showerror("Error occur","Please fill Every Parameter",parent=self.appu)
        elif self.txt_frame7.get()!=self.txt_frame8.get():
            messagebox.showerror("Password Error","Comfirmed Password is not same",parent=self.appu)
        elif self.chk_check.get()==0:
            messagebox.showerror("ConditionError","Condition is not satisfy",parent=self.appu)
        else:
            try:
                con = pymysql.connect(host="localhost",user="root",password="",database="people")
                cur = con.cursor()
                cur.execute("select * from people where email=%s",self.txt_frame4.get())
                row = cur.fetchone()
                #print(row)
                if row!=None:
                    messagebox.showerror("Error","Already enter the data with same email id",parent=self.appu)
                else:
                    cur.execute("insert into people (f_name,l_name,contact,email,question,answer,password) value(%s,%s,%s,%s,%s,%s,%s)",
                           (
                               self.txt_frame1.get(),
                               self.txt_frame2.get(),
                               self.txt_frame3.get(),
                               self.txt_frame4.get(),
                               self.txt_frame5.get(),
                               self.txt_frame6.get(),
                               self.txt_frame7.get()
                           ))
                    con.commit()
                    con.close()
                    messagebox.showinfo("Success","Register succefully",parent=self.appu)
                    self.clear()
            except Exception as es:
                messagebox.showerror("Error",f"Error due to : {str(es)}",parent=self.appu)
                
            
            
    
    #print(self.txt_frame1.get(), self.txt_frame2.get(),self.txt_frame3.get(),self.txt_frame4.get(),self.txt_frame5.get(),self.txt_frame6.get(),self.txt_frame7.get(),self.txt_frame8.get())        
        
        
```

```python
appu = Tk() 
obj  = register(appu)
appu.mainloop()
```
